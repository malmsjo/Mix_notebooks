```python
#Import functions
from __future__ import absolute_import, division, print_function, unicode_literals

# TensorFlow and Keras
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers
from tensorflow.keras.regularizers import l2
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Helper libraries
import random
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.utils import class_weight
import pandas as pd
from PIL import Image

from tensorflow.python.util import deprecation
deprecation._PRINT_DEPRECATION_WARNINGS = False
print(tf.__version__)
```

```python
#Load the dataset
def load_dataset():
    dirname = '/scratch-shared/bbbc021v1_images'
    x_orig = np.zeros((540, 256, 256, 3), dtype=np.float32)

    for f in range(x_orig.shape[0]):
        img    = Image.open(dirname + '/bbbc021v1_%s.png' % str(f)) #%s is every round in the loop in this case
        img    = np.array(img)
        x_orig[f] = img

    labels = pd.read_csv('/scratch-shared/bbbc021v1_labels.csv', #Read in the metadatafile
                          sep=";")
    y_orig = np.array(labels[:5]) #Ground truth

    return x_orig, y_orig #Return x_orig which is the matrix/array of the images, and y_orig which is the ground truth

print("Shape of X_orig = " + str(x_orig.shape))
print("Shape of y_orig = " + str(y_orig.shape))
```

```python
#One hot encoding
#One hot encoding is a process by which categorical variables are converted into a form that 
#could be provided to ML algorithms to do a better job in prediction (makes categories binary)

def convert_to_one_hot(y, C):
    moa_dict = {'Compound': 0, 'Control': 1}

    y = np.asarray([moa_dict[item] for item in y])
    y = np.eye(C)[y]
    y = y.astype('float32')

    return y
```

```python
#Function for plotting of the history of the model
def plot_history(model_history, model_name):
    fig = plt.figure(figsize=(15, 5), facecolor='w')
    ax = fig.add_subplot(131)
    ax.plot(model_history.history['loss'])
    ax.plot(model_history.history['val_loss'])
    ax.set(title=model_name + ': Model loss', ylabel='Loss', xlabel='Epoch')
    ax.legend(['train', 'valid'], loc='upper right')
    
    ax = fig.add_subplot(132)
    ax.plot(np.log(model_history.history['loss']))
    ax.plot(np.log(model_history.history['val_loss']))
    ax.set(title=model_name + ': Log model loss', ylabel='Log loss', xlabel='Epoch')
    ax.legend(['Train', 'Test'], loc='upper right')    

    ax = fig.add_subplot(133)
    ax.plot(model_history.history['accuracy'])
    ax.plot(model_history.history['val_accuracy'])
    ax.set(title=model_name + ': Model accuracy', ylabel='Accuracy', xlabel='Epoch')
    ax.legend(['train', 'valid'], loc='upper right')
    plt.show()
    plt.close()
```

```python
#Plotting confusion matrix for evaluation of the classification
def plot_confusion_matrix(cm, classes, model_name,
                          cmap=plt.cm.Blues):
    title = model_name + ': Confusion Matrix'
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
#Read the dataset
X_orig, y_orig = load_dataset()

print("Shape of X_orig = " + str(X_orig.shape))
print("Shape of y_orig = " + str(y_orig.shape))
```

```python
#One hot encoding of targets and standarization of images
#Y = convert_to_one_hot(y_orig, 2) #convert to one hot encoding
X = X_orig/255. #standarization 
Y = X_orig/255

print("Shape of Y_orig = " + str(Y.shape))
print("Shape of X_orig = " + str(X.shape))
```

```python
#Split data into training and validation
n_train = 500

random.seed(5026) #?
indices = np.arange(len(Y)) 
random.shuffle(indices)

X_train, X_valid = X[indices[:n_train]], X[indices[n_train:]]
Y_train, Y_valid = Y[indices[:n_train]], Y[indices[n_train:]]
```

```python
#Plot samples of the data
# Run cell to plot few example images of the training set
fig, ax = plt.subplots(2, 3, figsize=(14, 10))
fig.suptitle("Examples of training images", fontsize=20)
axes = ax.ravel()
for i in range(len(axes)):
    idx = np.where(np.argmax(Y_train, axis=1) == i)[0]
    s_idx = np.random.choice(idx)
    img = (X_train[s_idx]*255).astype("uint8")
    #axes[i].set_title(y_orig[indices][s_idx], fontsize=14)
    axes[i].imshow(img)
    axes[i].set_axis_off()
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
plt.close()
```

```python
#Define and summarize the LeNet-model

# define the model
num_classes = 2

inps = layers.Input((256, 256, 3)) # input image dimensions

# convolution and pooling layers
x = layers.Conv2D(16, kernel_size=(5, 5), strides=2, padding='same', activation='relu')(inps)
x = layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Conv2D(32, kernel_size=(5, 5), strides=1, padding='same', activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(4, 4))(x)
x = layers.Conv2D(64, kernel_size=(5, 5), strides=1, padding='same',activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(4, 4))(x)

# fully connected layers
x = layers.Flatten()(x)
x = layers.Dense(1024, activation='relu')(x)
x = layers.Dropout(0.2)(x)
preds = layers.Dense(num_classes, activation='softmax')(x)

LeNet = models.Model(inputs=inps, outputs=preds)
LeNet.summary()
```

```python
#Compile the LeNet-model
lr = 0.001
LeNet.compile(optimizer=optimizers.Adam(lr=lr), 
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Fit and evaluate the LeNet model
batch_size = 32
n_epochs = 30

history = LeNet.fit(X_train, Y_train,
                    batch_size=batch_size,
                    epochs=n_epochs,
                    validation_data=(X_valid, Y_valid),
                    verbose=2)

plot_history(history, 'LeNet')
valid_evaluate(LeNet, 'LeNet')
```
